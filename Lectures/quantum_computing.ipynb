### Generating a Uniform Distribution
For n qubits we have $2^{n}$ values, so a discrete distribution over $2^{n}$ points. 

$H^{\otimes n} \ket{0^{\otimes n}} = (H\ket{0}) \otimes \cdots \otimes (H\ket{0})$

$H^{\otimes n} \ket{0^{\otimes n}} = \frac{1}{2^{n/2}} \sum_{i=0}^{2^{n}-1} \ket{i}$


```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

qc = QuantumCircuit(1)
qc.h(0)
qc.measure_all()
qc.draw('mpl')
```

```python
backend = AerSimulator()
shots = 1000
compiled = transpile(qc, backend)
result = backend.run(compiled, shots=shots).result()
plot_histogram(result.get_counts())
```

for 6 qubits now

```python
n = 6 
qc = QuantumCircuit(n)
for i in range(n):
    qc.h(i)
qc.measure_all()
qc.draw('mpl')
```

```python
backend = AerSimulator()
shots = 50000
results = backend.run(transpile(qc, backend), shots=shots).result()
plot_histogram(results.get_counts())
```
